```python
# import libraries
from bs4 import BeautifulSoup
import numpy as np
from time import sleep
from random import randint
from selenium import webdriver
```

```python
def remove_tags(s):
    s = s.replace('<title>', '').replace('</title>', '').replace('<link>', '').replace('</link>', '')
    return s
```

```python
class itemClass:
    def __init__(self, title, link):
        self.title = title
        self.link = link
```

```python
#Loop to go over all pages
pages = np.arange(1, 3, 1)
data=[]

for page in pages:
    
    page="https://rss.app/feeds/XenZCYoo7eqBLeHy.xml" + str(page) 
    driver = webdriver.Chrome()
    driver.get(page)  
    sleep(randint(2,10))
    soup = BeautifulSoup(driver.page_source, 'xml')
    my_table = soup.find_all(['item'])

    for item in my_table:
        title = item('title')
        link = item('link')

        obj = itemClass(remove_tags(str(title)), remove_tags(str(link)))
        data.append(obj)

    # for tag in my_table:
    #     data.append(tag.get_text())

# for obj in data:
#     print(obj.title)
```

```python
import json

json_array = json.dumps([{'title': item.title, 'link': item.link} for item in data], indent=2)

# Print the JSON array
print(json_array)
```
